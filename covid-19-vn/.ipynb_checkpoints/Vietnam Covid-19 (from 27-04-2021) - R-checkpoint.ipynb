### Prepare data

```r
library(httr)
library(tidyverse)
library(rethinking)
```

```r
url_by_map <- 'https://vnexpress.net/microservice/sheet/type/covid19_2021_by_map'
url_by_total <- 'https://vnexpress.net/microservice/sheet/type/covid19_2021_by_total'

data_by_map <- rawToChar(GET(url_by_map)$content)
data_by_total <- rawToChar(GET(url_by_total)$content)
```

```r
write.csv(read.csv(text=data_by_map), './data_by_map.csv', row.names=TRUE)
```

```r
write.csv(read.csv(text=data_by_total), './data_by_total.csv', row.names=TRUE)
```

```r
data_by_map <- tibble(read.csv('./data_by_map.csv'))
data_by_total <- tibble(read.csv('./data_by_total.csv'))
```

```r
glimpse(data_by_map)
```

```r
print(
    local_data <- data_by_map %>% 
        rename("province" = "TỈNH.THÀNH", population="DÂN.SỐ..người.", area="DIỆN.TÍCH..km2.") %>%
        mutate(population = as.numeric(gsub("\\.", "\\", population))) %>%
        select(province, population, area) %>%
        drop_na()
)
```

```r
glimpse(data_by_total)
```

The code below will tidy the data above to more useablility format.

```r
print(
    cases_by_province <- data_by_total %>%
        select(-X, -X.1, -X.2) %>%
        rename_with(~gsub('\\.', '\\ ', .)) %>%
        rename("Bà Rịa - Vũng Tàu" = "Bà Rịa   Vũng Tàu") %>%
        mutate(Ngày = as.Date(Ngày, tryFormats = "%d/%m")) %>%
        filter_all(any_vars(. != 0)) %>%
        drop_na()
)
```

```r
library(reshape2)
options(repr.plot.width = 10, repr.plot.height = 6)

cases_by_province %>%
    pivot_longer(c(everything(), -Ngày), names_to = "province", values_to = "cases") %>%
    ggplot(aes(x = `Ngày`, y = cases, group = province, color = province)) + geom_smooth()
```

```r
options(repr.plot.width = 12, repr.plot.height = 10)

local_data %>%
    ggplot(aes(x = reorder(province, population), y = population, color = province)) + 
        geom_bar(stat="identity", show.legend = FALSE) + 
        geom_text(aes(label=population), hjust="right", color="white", size=3, position = position_dodge(.9)) +
        coord_flip() + 
        theme_minimal()
```
