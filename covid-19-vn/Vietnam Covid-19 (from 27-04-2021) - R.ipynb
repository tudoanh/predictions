```r
library(httr)
library(tidyverse)
library(rethinking)
```

```r
url_by_map <- 'https://vnexpress.net/microservice/sheet/type/covid19_2021_by_map'
url_by_total <- 'https://vnexpress.net/microservice/sheet/type/covid19_2021_by_total'

data_by_map <- rawToChar(GET(url_by_map)$content)
data_by_total <- rawToChar(GET(url_by_total)$content)
```

```r
write.csv(read.csv(text=data_by_map), './data_by_map.csv', row.names=TRUE)
```

```r
write.csv(read.csv(text=data_by_total), './data_by_total.csv', row.names=TRUE)
```
